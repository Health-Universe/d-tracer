```python
import pandas as pd
import numpy as np
```

```python
data = '../data/2022_04_22_NR_lipidomics_DAARSL3_HILIC_NEG_DAAvsRSL3DAA.csv'
### Need a way for .py file to ask which file to upload
```

```python
df = pd.read_csv(data,header=2)
df.head()
```

```python
'''Create a list of all column names'''
col = df.columns.tolist()
print(col)

'''Index of columns we wish to keep and analyze'''
keep = [0,2,4,5]
# to avoid errors, we should change this to sort by column name instead of index

'''Column names we wish to keep and analyze'''
col_keep = []
for i in keep:
    col_keep.append(col[i])
col_keep
```

```python
'''Make a df of sample mz, rt, ccs, and intensities'''
intensities = col[16:24:1]
df_int = df[intensities][:]
df_int.head()
```

```python
'''Create new filtered dataframe of important columns'''
df_keep = df[col_keep][:]
df_keep
```

```python
'''Define lists and tolerances of each column to compare with itself'''
mz = df_keep['m/z']
mz_tol = 0.005
rt = df_keep['Retention time (min)']
rt_tol = 1e-3
ccs = df_keep['CCS (angstrom^2)']
ccs_tol = .03

'''Initialize a list for indexes to be held for each pair of matched values'''
idxs = []
'''Create nested for loop to compare i with j in each column'''
for i in range(len(df_keep)):
    for j in range(len(df_keep)):
        '''Define checks for each column'''
        check_mz = np.isclose(mz[i], mz[j]+6.03, atol = mz_tol)        
        check_rt = np.isclose(rt[i], rt[j], rt_tol)
        check_ccs = np.isclose(ccs[i], ccs[j], rtol = ccs_tol)
        
        '''Record results of each check (True/False)'''
        checks = [check_mz, check_rt, check_ccs]
        
        '''If all checks are true, append to list of pairs'''
        if all(checks) and i!=j:
            idxs.append([i,j])
            pairs = np.array(idxs)
        else:
            pass
print(pairs)
```

```python
for i in range(len(idxs)):
    print(df_keep.iloc[pairs[i][:]])
### Not sure what we want the end result to look like
```

```python
'''add matching indicies to list and make new df of matches with mz, rt, ccs'''
idx_list = []
for i in pairs:
    for j in i:
        idx_list.append(j)
df_pairs = df_keep.loc[df_keep.index[idx_list]]
df_pairs.head()
```

```python
'''combine mz, rt, ccs df with intensities df'''
df_pair_int = df_int.loc[df_int.index[idx_list]]
df_output = pd.concat([df_pairs, df_pair_int], axis="columns")
df_output.head()
```

```python
'''export to excel file'''
df_output.to_excel('../data/2022_04_22_NR_lipidomics_DAARSL3_HILIC_NEG_DAAvsRSL3DAA_pairs.xlsx')
```

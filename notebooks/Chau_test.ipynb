```python
import pandas as pd
import numpy as np
import time
import sys

data = '../data/2022_04_22_WT_DAARSL3_NEG_allcompunds.csv'
standard = '../data/2022_04_22_NEG_all_standards_abundance.csv'

def upload(file, limit=None):
    df = pd.read_csv(file, header=2, nrows=limit)
    return df

df_raw = upload(data)
df_standard_raw = upload(standard)
# df_raw
```

```python
def format(df, samples=None):
    col_main = ['Compound',
                'm/z',
                'Retention time (min)',
                'CCS (angstrom^2)']
    df_keep = df[col_main].sort_values(by=["Compound"], ascending=False).reset_index(drop=True)
    return df_keep
df_keep = format(df_raw)
df_keep
```

```python
def id_standards(df_keep, df_standard_keep):

    check_rt = []
    for i, row in df_standard_keep.iterrows():
        mz_standard = row['m/z']
        rt_standard = row['Retention time (min)']
        check_mz = df_keep[np.isclose(df_keep['m/z'], mz_standard)]
        if check_mz.shape[0] < 1:
            print (f"Warning: Cannot find matching standards {i}")
        else: continue 
        check_rt = df_keep[np.isclose(df_keep['Retention time (min)'], rt_standard)]
        if check_rt.shape[0] < 1:
            print (f"Warning: Cannot find matching standards {i}")
        else: continue 

    print(check_rt)
    return check_rt 


```
